# 3 Visualizations

### 3.1 Line Plot

USE: to show trends, developments, or changes over time.

EXAMPLE: Daily, Weekly, monthly and annual company sales.

DATA: Numerical data over time.

Covid 19 (Third dataframe) is the best suited for a line plot because it shows the trend over time

```python
# importing visualizations libraries in python
import seaborn as sns    #for beautifull visualizations
import plotly.express as px  #for interactive visualizations
import matplotlib.pyplot as plt #for normal visualizations
%matplotlib inline
```

### 3.1.1 Line plot using seaborn

```python
#pick data for a specific country (in this case Kenya) and then slice the dates.
kenyan_data = third_df[third_df['Country/Region'] == 'Kenya']
date_columns = kenyan_data.columns[4:]

# Melt the DataFrame to make it suitable for visualization by transposing it.
melt_df = kenyan_data.melt(id_vars=['Country/Region'], value_vars=date_columns, var_name='Date', value_name='Cases of Covid-19')

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Cases of Covid-19', data=melt_df)
plt.xticks(rotation=90)
plt.title('Seaborn Line Plot for Kenya  Covid-19 Cases')
plt.show()
```

Because the date labels are not visible we will group the dates

```python
#group the data
melt_df['Date'] = pd.to_datetime(melt_df['Date'])
weekly_data = melt_df.groupby(pd.Grouper(key='Date', freq='W-Mon'))['Cases of Covid-19'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Cases of Covid-19', data=weekly_data)
plt.xticks(rotation=90)
plt.title('Seaborn Line Plot for Kenya Covid-19 Cases Grouped')
plt.show()
```

### 3.1.2 Line plot using matplotlib

```python
#line plot using matplotlip
plt.figure(figsize=(12, 6))
plt.plot(weekly_data['Date'], weekly_data['Cases of Covid-19'])
plt.xticks(rotation=45,) 
plt.xlabel('Date')
plt.ylabel('Cases of Covid-19')
plt.title('Matplotlib Line Plot for Kenya Covid-19 Cases Grouped')
plt.show()
```

### Line plot using plotly

```python
#line plot using plotly
fig = px.line(melt_df, x='Date', y='Cases of Covid-19', title='Plotly Line Plot for Kenya Covid-19 Cases')
fig.show()
```

### 3.2 Pie Chart

USE:  to represent parts of a whole.

EXAMPLE: percentage of different types of food crops grown on a farm.

DATA: Categorical data.

In our cases we will use the movies dataset to show the type of movie


### 3.2.1 Pie Chart using matplotlib

```python
#pie chart using matplotlib
movie_type = fourth_df["type"].value_counts()

plt.figure(figsize=(12,6))
plt.pie(movie_type, labels=movie_type.index,autopct='%1.2f%%')
plt.title('Matplotlib Pie Chart for type of movie')
plt.show()
```

### 3.2.2 Pie Chart using plotly

```python
#pie chart using plotly
fig = px.pie(values=movie_type, names=movie_type.index, title="Plotly Pie Chart for type of movie")
fig.show()
```

>> Pie charts are available only available in plotly and matplotlib. But seaborn does not have a similar function but it has an almost similar plot called count plot. 

```python
# Using countplot
sns.countplot(data=fourth_df, x="type", )
plt.title('Seaborn type of movie')
plt.show()
```

### 3.3 Bar Chart

USE: to show the comparison of different categories.

EXAMPLE: number of students and their preferred meal.

DATA: Categorical or numeric data in categories.

In our case we will use the movie dataset and plot runtime against age certification


### 3.3.1 Bar plot using  matplotlib

```python
# plotting a bar graph using matplotlib
plt.figure(figsize=(12, 6))
plt.bar(fourth_df['age_certification'].astype(str), fourth_df['runtime'])
plt.xlabel('Age Certification')
plt.ylabel('Runtime')
plt.title('Matplotlib Bar Plot of Runtime by Age Certification')
plt.show()
```

### 3.3.2 Bar plot using Seaborn

```python
#Barplot using Seaborn 
plt.figure(figsize=(12, 6))
sns.barplot(x='age_certification', y='runtime', data=fourth_df)
plt.xlabel('Age Certification')
plt.ylabel('Runtime')
plt.title('Seaborn Bar Plot of Runtime by Age Certification')
plt.show()
```

### 3.3.3 Bar plot using Plotly

```python
# Barplot using plotly
fig = px.bar(fourth_df, x='age_certification', y='runtime', color='age_certification', 
             title='Plotly Bar Plot of Runtime by Age Certification', color_discrete_sequence=px.colors.qualitative.Set1)
fig.update_layout(xaxis_title='Age Certification', yaxis_title='Runtime')
fig.show()
```

### 3.4 Histogram
USE: to show how frequently numerical values occur.

EXAMPLE: Distribution of different age groups in a country.

DATA: Numeric data representing one variable.

in our case apple quality is the best to use for histogram

### 3.4.1 Histogram using matplotlib

### 3.6 Heat Maps.

USE: to represent values by use of colors.

EXAMPLE: Most preferred region to invest.

DATA: Numeric data

In our case we will use the apple quality data



### 3.6.1 Heat Maps using seaborn

```python
#pick only numerical data
heatmap_df = second_df[['Juiciness', 'Ripeness', 'Size', 'Weight', 'Sweetness', 'Crunchiness']]

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Variable Correlations')
plt.show()
```

Plotly and matplotlib do not have a direct way to plot  but you can use

### 3.7 Box-and-Whisker Plots

USE: to provide a numeric summary of variables visually, it can also be used to detect outliers.

EXAMPLE: Detecting the outlier value of student performance.

DATA: Numeric data but for a single variable.

In our case we will consider the sweetness variable from the apple quality dataset


### 3.7.1 Box-and-Whisker Plots using Matplotlib

```python
# Box-and-Whisker plot using matplotlib
plt.figure(figsize=(12, 6))
plt.boxplot(second_df['Sweetness'].dropna())
plt.title('Matplotlib Box and Whisker Plot for Sweetness')
plt.xlabel('Sweetness')
plt.show()
```

### 3.7.2 Box-and-Whisker plot using seaborn

```python
#Box-and-whisker plot using seaborn
plt.figure(figsize=(12, 6))
sns.boxplot(data=second_df, x='Sweetness')
plt.title('Seaborn Box and Whisker Plot for Sweetness')
plt.show()
```

### 3.7.3 Box and whisker plot using plotly

```python
#Box and whisker plot using plotly
fig = px.box(second_df, x='Sweetness', title='Plotly Box and Whisker Plot for Sweetness')
fig.show()
```

### 3.8 Word clouds

USE: to show the occurrence and distribution of text data.

EXAMPLE: Most occurring words in a book or comment section.

DATA: Text data

in our case we will consider the Amazon 2024 valentines dataset


```python
from wordcloud import WordCloud

text = ' '.join(first_df['description'].dropna())
wordcloud = WordCloud(background_color='white').generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Product Descriptions')
plt.show()
```
